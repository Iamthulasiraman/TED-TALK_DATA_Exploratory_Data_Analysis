### adding month and year columns in dataframe

```python
month = []
year  = []
for i in df['date']:
    year.append(i[-4:])
    month.append(i[0:-4])
df['month'] = month 
df['year'] = year
```

```python
df.head()
```

### month-wise analysis of ted talks frequency

```python
df['month'].unique()
    
```

```python
list = []
for i in df['month'].unique():
    list.append([i,len(df[df['month'] == i])])
```

```python
list
```

### year-wise analysis of ted talks frequency
